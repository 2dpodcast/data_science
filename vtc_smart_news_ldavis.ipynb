# LDAvis on VNE corpus


- How well topic - term related?
- Distinctiveness vs saliency
- Topic distances

```python
import gensim

dictionary = gensim.corpora.Dictionary.load('/home/laampt/nlp/xnews/model2/xnews.280Kdoc.dict.vne')
corpus = gensim.corpora.MmCorpus('/home/laampt/nlp/xnews/model2/xnews.tfidfcorpus.mm.vne')
lda = gensim.models.ldamodel.LdaModel.load('/home/laampt/nlp/xnews/model2/xnews.lda.mm2.vne')
```

```python
print dictionary
print corpus
print lda
```

```python
print '\n'.join(lda.print_topics())
```

```python
import pyLDAvis.gensim as gensimvis
import pyLDAvis
```

```python
%%time

vis_data = gensimvis.prepare(lda, corpus, dictionary)
```

```python
from sklearn.externals import joblib
```

```python
joblib.dump(vis_data, "/home/laampt/nlp/xnews/model2/vis_data.pkl.vne")
```

```python
pyLDAvis.display(vis_data)
```
